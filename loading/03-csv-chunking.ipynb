### 2. 파싱 + 청킹
- 공식문서 : https://docs.langchain.com/oss/python/integrations/document_loaders/unstructured_file
- `langchain-text-splitters` 패키지의 `RecursiveCharacterTextSplitter`를 사용하여 chunking을 할 수도 있겠지만 이 방식은 `글자수로 자르는 방식`이여서 CSV 파일에는 적절치 않기 때문에 `langchain_unstructured` 패키지의 `UnstructuredLoader` 클래스 내부 옵션을 활용해서 사용함
- UnstructuredLoader(파일명, 옵션..).`load_and_split()` 방식은 `Deprecated` 되었으므로 사용하지 말아야하기 때문에(https://reference.langchain.com/python/langchain-core/document_loaders/base/BaseLoader/load_and_split) `공식문서 권장 방식`인 UnstructuredLoader(파일명, 옵션에서 청킹 설정).`load()` 방식으로 진행함

```python
from langchain_unstructured import UnstructuredLoader

# 옵션 설명 #
    # 첫번째 옵션위치: 파일경로 + 파일명
    # chunking_strategy: 청킹전략 - 요소(element)들을 어떻게 청크로 묶을지 결정
        # basic 전략: 구조 무시, 크기(max_characters) 채워서 자르기
        # by_title 전략: 제목(섹션) 만나면 새 청크 시작 (구조 있는 문서에 유리)
    # max_characters: 한 청크의 최대 글자 수(하드 리밋, 절대 못 넘음). 값이 클수록 덜 잘림
    # new_after_n_chars: 소프트 리밋 - 이 글자 수를 넘으면 청크를 미리 끊음(max_characters 이하에서 조기 분할)
    # include_orig_elements: 청크에 합쳐지기 전 원본 요소를 메타데이터로 남길지 (False=버려서 가볍게)
    # overlap: 겹칠 글자 수. (주의) 기본적으로는 '하나의 큰 element가 max_characters를 넘어 강제 분할될 때'만 겹침이 적용됨
    # overlap_all: True로 주면 위 경우뿐 아니라 '모든 인접 청크 경계'에 overlap을 적용 (문맥 끊김 방지 강화, 단 중복↑·표 등에 부작용 주의)
    # (참고) encoding: 현재 파일은 UTF-8이라 생략. cp949 원본을 다시 받아 쓰면 encoding='cp949' 추가 필요
loader = UnstructuredLoader(
    'files/hospital-2025-11.csv',
    chunking_strategy='basic',
    max_characters=1000,
    new_after_n_chars=800,
    include_orig_elements=False,
    overlap=200,
    # overlap_all=True,   # 모든 청크 경계에 overlap 적용하려면 주석 해제
)
```

```python
docs = loader.load()
```
