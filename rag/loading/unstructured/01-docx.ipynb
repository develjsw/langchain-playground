### 1. 패키지 다운로드
- langchain-unstructured : 어댑터(인터페이스)
- unstructured : 실제 파싱 엔진

unstructured는 여러가지 파일을 파싱할 수 있는 기능 전체가 들어가 있기 때문에 무거움

```python
%pip install -qU langchain langchain-unstructured unstructured
```

### 2. 파싱

```python
from langchain_unstructured import UnstructuredLoader

docs = UnstructuredLoader(
    '../../files/doc/tax.docx'
).load()
```

```python
docs
```
